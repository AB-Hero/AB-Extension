```python
from lxml import html
import requests
```

```python
class node(object):
    def __init__(self, value, children = []):
        text = value.text
        if text == None:
            text = ""

        if value.tag == None:
            self.value = "None"
        else:
            self.value = str(value.tag)      

        self.value = self.value +  ", text: " + text
        self.children = [node(i) for i in value]
    def __str__(self, level=0):
        ret = "\t"*level+repr(self.value)+"\n"
        for child in self.children:
            ret += child.__str__(level+1)
        return ret

    def __repr__(self):
        return '<tree node representation>'
```

```python
res = requests.get("https://boxbox.in/products/copy-of-boxbox-oversized-t-shirt-44-lewis-hamilton-1")
tree = html.fromstring(res.text)
```

```python
n = node(tree, children =[i for i in tree])
print(n)
```
