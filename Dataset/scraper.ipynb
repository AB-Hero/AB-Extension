# Create DOM Tree

```python
# change url for corresponding webpage
def createTree(url):
    session = HTMLSession()
    r = session.get(url)
    tree = html.fromstring(r.text)
    soup = BeautifulSoup(r.text, "html.parser")
    return soup,tree, r.text

def findProductLinks(soup):
    links = []
    for link in soup.find_all('a', href=True):
        if 'shop/product' in link['href']:
            links.append(link['href'])

    return links
    


```

```python
nodes = {}
code = {}
urls = {}
trees = {}

for link in webUrls:
    print("scraping product page for {} ".format(link))
    s,t,c = createTree(link)
    links = findProductLinks(s)
    del links[0]

    for prodLink in links:
        l = "https://macys.com" + prodLink
        nodes[l] = {}
        ps, pt, pc = createTree(l)
        code[l] = pc
        trees[l] = pt

        for section in webUrls[link]:
            nodes[l][section] = []

            for params in webUrls[link][section]:    
                tag = params[0]
                attrib = params[1]
                val = params[2]
                query = "//{}[@{} = '{}']".format(tag, attrib, val)
                found = pt.xpath(query)
                nodes[l][section].extend(found) 
```

# Create labels for nodes

```python
""""
nodes are for one url only
nodes is a dictionary with sections for keys, and list of nodes for values
nodes = {"navigation": [node1, node2], "product-info": [node3, node4] ...}

"""
def labelNodes(nodeDict, tree):
    labels = {"Navigation": 0, 
              "Product Info": 1, 
              'Product Versions': 2, 
              "Product Images": 3, 
              'Checkout Options': 4, 
              'Shopping Cart' : 5,
              "Other": 6}
    nodeIds = {}
    labs = []

    count = 0
    for i in tree.getiterator():
        nodeIds[i] = count
        count += 1
        labs.append(labels["Other"])

    
    root = tree.getroottree()
    allNodes = []
    for section in nodeDict:
        for n in nodeDict[section]:
            path = root.getpath(n)
            l = len(path.split('/'))
            allNodes.append((l, n, section))

    allNodes = sorted(allNodes, key= lambda x: x[0])

    def labelChildren(node, section):
        children = node.getchildren()
        id = nodeIds[node]
        labs[id] = labels[section]

        for i in children:
            labelChildren(i, section)

    for node in allNodes:
        labelChildren(node[1], node[2])

    return labs
    
    
```

```python
fileName = "data.json"

data = {"urls": [], "code": {}, "labels": {}}
urls = []
labels = {}


for i in nodes:
    urls.append(i)

for i in nodes:
    # print(trees[i].)

    labels[i] = labelNodes(nodes[i], trees[i])



data['urls'] = urls
data['code'] = code
data['labels'] = labels


j = json.dumps(data)
f = open(fileName, 'w')
f.write(j)
f.close()







    

        
```
